```python
import pandas as pd
from statsmodels.graphics.tsaplots import plot_pacf
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('data/Account Products.xlsx', sheet_name='web_traffic_accounts', header=0)
data.head()
```

```python
# set date as index 
data.set_index('Date', inplace=True)
data.head()
```

```python
time_series = data['Web Traffic']
```

```python
plot_pacf(data['abanca.com/es/cuentas/cuenta-online'].diff().dropna(), lags=12, alpha=0.01)
plt.title(f'PACF')
plt.ylabel('Correlation')
plt.xlabel('Lags')
plt.show()
```

```python
dates = df.iloc[0, 2:]
dates
```

```python
# Extract bank names and categories
bank_names = df.iloc[0:, 1]  # Bank names are in the second column starting from the second row
categories = df.iloc[0:, 0]  # Categories are in the first column for each bank

bank_names, categories
```

```python
filled_categories = categories.fillna(method='ffill')
filled_categories 
```

```python
# Create a DataFrame by combining the two series
combined_df = pd.DataFrame({
    'Category': filled_categories,
    'Bank Name': bank_names
})

combined_df
```

```python
category_counts = combined_df.groupby('Bank Name')['Category'].nunique()

import matplotlib.pyplot as plt

# Plotting the number of unique categories for each bank
category_counts.plot(kind='bar', figsize=(14, 6), color='skyblue')
plt.title('Number of Unique Categories per Bank')
plt.xlabel('Bank Name')
plt.ylabel('Number of Unique Categories')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlap
plt.show()
```

### Grouping by Category, and count unique number of category for each banks. Also less length of bank account name

```python
category_counts = combined_df.groupby(['Bank Name', 'Category']).size().unstack(fill_value=0)

# Plotting using a stacked bar chart
category_counts.plot(kind='bar', stacked=True, figsize=(14, 6), color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Number of Categories per Bank')
plt.xlabel('Bank Name')
plt.ylabel('Count of Categories')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlap
plt.legend(title='Category')
plt.show()
```

```python
category_counts = combined_df.groupby(['Bank Name', 'Category']).size().unstack(fill_value=0)
unique_categories = combined_df['Category'].unique()

for category in unique_categories:
    # Select the banks where the count in the current category is greater than 1
    filtered_banks = category_counts[category_counts[category] == 1]

    # Check if the filtered DataFrame is empty; if not, plot it
    if not filtered_banks.empty:
        # Plotting the number of unique categories for banks that have more than one entry in this category
        ax = filtered_banks.plot(kind='bar', stacked= True, figsize=(10, 6), title=f'Number of categories of banks ranked in "{category}" ',color=['skyblue', 'lightgreen', 'salmon'])
        ax.set_xlabel('Bank Name')
        ax.set_ylabel(f'Count of "{category}"')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

```python
bank_counts = combined_df.groupby('Category')['Bank Name'].nunique()

import matplotlib.pyplot as plt

# Plotting the number of unique categories for each bank
bank_counts.plot(kind='bar', figsize=(14, 6), color='skyblue')
plt.title('Number of Unique Bank per Account')
plt.xlabel('Account Type')
plt.ylabel('Number of Unique Bank')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlap
plt.show()
```

### Bubble plot

```python

# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Create bubble charts for each category
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    bank_names = category_data['Bank Name'].unique()
    mean_ranks = []
    variances = []
    filtered_bank_names = []
    
    for bank in bank_names:
        bank_data = category_data[category_data['Bank Name'] == bank]
        if not bank_data['Rank'].isnull().all():
            mean_ranks.append(bank_data['Rank'].mean())
            variances.append(bank_data['Rank'].var())
            filtered_bank_names.append(bank)
    
    plt.figure(figsize=(12, 8))
    bubbles = plt.scatter(filtered_bank_names, mean_ranks, s=np.array(variances)*300, alpha=0.5)
    
    plt.title(f'Rank Position Variance for {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Mean Rank Position')
    plt.xticks(ticks=range(len(filtered_bank_names)), labels=filtered_bank_names, rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### Box plot

```python
# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Create box plots for each category
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    bank_names = []
    rank_data = []
    
    for bank in category_data['Bank Name'].unique():
        bank_data = category_data[category_data['Bank Name'] == bank]
        ranks = bank_data['Rank'].dropna()
        if not ranks.empty:
            rank_data.append(ranks.values)
            bank_names.append(bank)
    
    plt.figure(figsize=(12, 8))
    box = plt.boxplot(rank_data, labels=bank_names, showmeans=True, meanline=True, patch_artist=True)
    
    # Customizing the box plot
    for median in box['medians']:
        median.set(color='orange', linewidth=2)
    
    for mean in box['means']:
        mean.set(marker='^', color='green', markersize=10)
    
    # Add observation count text
    for i, (rank, bank) in enumerate(zip(rank_data, bank_names)):
        plt.text(i + 1, max(rank) + 0.5, f'n={len(rank)}', horizontalalignment='center', fontsize=10, color='blue')
    
    plt.title(f'Rank Position Distribution for {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Rank Position')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Create box plots for each category
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    bank_names = []
    rank_data = []
    observation_counts = []
    
    for bank in category_data['Bank Name'].unique():
        bank_data = category_data[category_data['Bank Name'] == bank]
        ranks = bank_data['Rank'].dropna()
        if not ranks.empty:
            rank_data.append(ranks.values)
            bank_names.append(bank)
            observation_counts.append(len(ranks))
    
    # Combine data into a single list for sorting
    combined_data = list(zip(bank_names, rank_data, observation_counts))
    
    # Sort by observation count and then by mean rank within segments (ascending mean rank)
    combined_data.sort(key=lambda x: (x[2] < 5, 5 <= x[2] < 10, x[2] >= 10, np.mean(x[1])))
    
    # Unzip the combined data
    bank_names, rank_data, observation_counts = zip(*combined_data)
    
    plt.figure(figsize=(12, 8))
    box = plt.boxplot(rank_data, labels=bank_names, showmeans=True, meanline=True, patch_artist=True)
    
    # Customizing the box plot
    for median in box['medians']:
        median.set(color='orange', linewidth=2)
    
    for mean in box['means']:
        mean.set(marker='^', color='green', markersize=10)
    
    # Add observation count text
    for i, (rank, bank, obs_count) in enumerate(zip(rank_data, bank_names, observation_counts)):
        plt.text(i + 1, max(rank) + 0.5, f'n={obs_count}', horizontalalignment='center', fontsize=10, color='blue')
    
    # Add vertical lines for observation thresholds
    lt5_line_drawn = False
    lt10_line_drawn = False

    for i, count in enumerate(observation_counts):
        if not lt5_line_drawn and count < 5:
            plt.axvline(x=i + 0.5, color='red', linestyle='--', linewidth=2, label='< 5 observations')
            lt5_line_drawn = True
        if not lt10_line_drawn and count < 10:
            plt.axvline(x=i + 0.5, color='orange', linestyle='--', linewidth=2, label='< 10 and >= 5 observations')
            lt10_line_drawn = True
        if count >= 10 and i == len(observation_counts) - 1:
            plt.axvline(x=i + 0.5, color='green', linestyle='--', linewidth=2, label='>= 10 observations')
    
    plt.title(f'Rank Position Distribution for {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Rank Position')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

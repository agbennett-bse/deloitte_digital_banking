```python
# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Create plots for each category
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    plt.figure(figsize=(10, 5))
    for bank in category_data['Bank Name'].unique():
        bank_data = category_data[category_data['Bank Name'] == bank]
        plt.plot(bank_data['Date'], bank_data['Rank'], marker='o', linestyle='-', label=bank)
    
    plt.title(f'Rank Position Over Time for {category}')
    plt.xlabel('Date')
    plt.ylabel('Rank Position')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Bubble plot

```python

# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Create bubble charts for each category
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    bank_names = category_data['Bank Name'].unique()
    mean_ranks = []
    variances = []
    filtered_bank_names = []
    
    for bank in bank_names:
        bank_data = category_data[category_data['Bank Name'] == bank]
        if not bank_data['Rank'].isnull().all():
            mean_ranks.append(bank_data['Rank'].mean())
            variances.append(bank_data['Rank'].var())
            filtered_bank_names.append(bank)
    
    plt.figure(figsize=(12, 8))
    bubbles = plt.scatter(filtered_bank_names, mean_ranks, s=np.array(variances)*300, alpha=0.5)
    
    plt.title(f'Rank Position Variance for {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Mean Rank Position')
    plt.xticks(ticks=range(len(filtered_bank_names)), labels=filtered_bank_names, rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### Box plot

```python
# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Create box plots for each category
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    bank_names = []
    rank_data = []
    
    for bank in category_data['Bank Name'].unique():
        bank_data = category_data[category_data['Bank Name'] == bank]
        ranks = bank_data['Rank'].dropna()
        if not ranks.empty:
            rank_data.append(ranks.values)
            bank_names.append(bank)
    
    plt.figure(figsize=(12, 8))
    box = plt.boxplot(rank_data, labels=bank_names, showmeans=True, meanline=True, patch_artist=True)
    
    # Customizing the box plot
    for median in box['medians']:
        median.set(color='orange', linewidth=2)
    
    for mean in box['means']:
        mean.set(marker='^', color='green', markersize=10)
    
    # Add observation count text
    for i, (rank, bank) in enumerate(zip(rank_data, bank_names)):
        plt.text(i + 1, max(rank) + 0.5, f'n={len(rank)}', horizontalalignment='center', fontsize=10, color='blue')
    
    plt.title(f'Rank Position Distribution for {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Rank Position')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python

# Get unique categories
unique_categories = transformed_df['Category'].unique()

# Define thresholds
thresholds = {
    '≥10': lambda x: x >= 10,
    '<10 and ≥5': lambda x: 5 <= x < 10,
    '<5': lambda x: x < 5
}

# Create box plots for each category and each threshold
for category in unique_categories:
    category_data = transformed_df[transformed_df['Category'] == category]
    
    for threshold_name, threshold_func in thresholds.items():
        bank_names = []
        rank_data = []
        
        for bank in category_data['Bank Name'].unique():
            bank_data = category_data[category_data['Bank Name'] == bank]
            ranks = bank_data['Rank'].dropna()
            if threshold_func(len(ranks)):
                rank_data.append(ranks.values)
                bank_names.append(bank)
        
        if not bank_names:
            continue  # Skip if no data for this threshold
        
        plt.figure(figsize=(12, 8))
        box = plt.boxplot(rank_data, labels=bank_names, showmeans=True, meanline=True, patch_artist=True)
        
        # Customizing the box plot
        for median in box['medians']:
            median.set(color='orange', linewidth=2)
        
        for mean in box['means']:
            mean.set(marker='^', color='green', markersize=10)
        
        # Add observation count text
        for i, (rank, bank) in enumerate(zip(rank_data, bank_names)):
            plt.text(i + 1, max(rank) + 0.5, f'n={len(rank)}', horizontalalignment='center', fontsize=10, color='blue')
        
        plt.title(f'Rank Position Distribution for {category} ({threshold_name})')
        plt.xlabel('Bank Name')
        plt.ylabel('Rank Position')
        plt.xticks(rotation=45, ha='right')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd

file_path = 'D:/Hangze Wu/BSE/DSDM/semester3/Master Thesis/thesis project/data/Account Products.xlsx'
df = pd.read_excel(file_path, sheet_name='ranking_position', skiprows=7)

display(df.head())
```

```python
dates = df.iloc[0, 2:]
dates
```

```python
# Extract bank names and categories
bank_names = df.iloc[0:, 1]  # Bank names are in the second column starting from the second row
categories = df.iloc[0:, 0]  # Categories are in the first column for each bank

bank_names, categories
```

```python
filled_categories = categories.fillna(method='ffill')
filled_categories 
```

```python
# Create a DataFrame by combining the two series
combined_df = pd.DataFrame({
    'Category': filled_categories,
    'Bank Name': bank_names
})

combined_df
```

```python
category_counts = combined_df.groupby('Bank Name')['Category'].nunique()

import matplotlib.pyplot as plt

# Plotting the number of unique categories for each bank
category_counts.plot(kind='bar', figsize=(14, 6), color='skyblue')
plt.title('Number of Unique Categories per Bank')
plt.xlabel('Bank Name')
plt.ylabel('Number of Unique Categories')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlap
plt.show()
```

### Grouping by Category, and count unique number of category for each banks. Also less length of bank account name

```python
total_categories_per_bank = combined_df.groupby('Bank Name')['Category'].nunique()
```

```python
# Create a DataFrame that maps banks to categories
category_mapping = combined_df.drop_duplicates(subset=['Bank Name', 'Category'])

# Get a list of all unique categories
unique_categories = combined_df['Category'].unique()

# Plot for each category
for category in unique_categories:
    # Filter banks that are involved in the current category
    banks_in_category = category_mapping[category_mapping['Category'] == category]['Bank Name']
    
    # Filter the total_categories_per_bank to include only these banks
    filtered_counts = total_categories_per_bank[banks_in_category]
    
    # Plot
    plt.figure(figsize=(14, 6))
    filtered_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Number of Categories per Bank in {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Number of Unique Categories')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
category_mapping = combined_df.drop_duplicates(subset=['Bank Name', 'Category'])

# Get a list of all unique categories
unique_categories = combined_df['Category'].unique()

colors = plt.cm.tab20.colors  # Using a matplotlib colormap that provides 20 unique colors
color_map = dict(zip(unique_categories, colors[:len(unique_categories)]))

# Plot for each category
for category in unique_categories:
    # Filter banks that are involved in the current category
    banks_in_category = category_mapping[category_mapping['Category'] == category]['Bank Name']
    
    # Filter the total_categories_per_bank to include only these banks
    filtered_counts = total_categories_per_bank[banks_in_category]
    
    # Plot using the color mapped to the current category
    plt.figure(figsize=(14, 6))
    filtered_counts.plot(kind='bar', color=color_map[category])
    plt.title(f'Number of Categories per Bank in {category}')
    plt.xlabel('Bank Name')
    plt.ylabel('Number of Unique Categories')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
category_counts = combined_df.groupby(['Bank Name', 'Category']).size().unstack(fill_value=0)

import matplotlib.pyplot as plt

# Plotting using a stacked bar chart
category_counts.plot(kind='bar', stacked=True, figsize=(14, 6), color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Number of Categories per Bank')
plt.xlabel('Bank Name')
plt.ylabel('Count of Categories')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlap
plt.legend(title='Category')
plt.show()
```
